```python
import os
cd = os.getcwd()
cd
```

```python
avi_file_path = os.path.join(cd,'race_car_out.avi')
mp4_file_path = os.path.join(cd,'race_car_out.mp4')
```

```python
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

out_avi = cv2.VideoWriter(avi_file_path, cv2.VideoWriter_fourcc('M','J','P','G'), 10, (frame_width,frame_height))
out_mp4 = cv2.VideoWriter(mp4_file_path, cv2.VideoWriter_fourcc(*'XVID'), 10, (frame_width,frame_height))
```

```python
while(cap.isOpened()):
  ret, frame = cap.read()
  if ret == True:
    print('writing into files')
    out_avi.write(frame)
    out_mp4.write(frame)
  else:
    print('not writing into files')
    break
```
