```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
%matplotlib inline
from IPython.display import Image
```

```python
img_bgr = cv2.imread('nz_coast.jpg',cv2.IMREAD_COLOR)
img_rgb = cv2.cvtColor(img_bgr,cv2.COLOR_BGR2RGB)
Image(filename='nz_coast.jpg')
```

```python
#Addition or Brightness
matrix = np.ones(img_rgb.shape, dtype="uint8")*50
img_rgb_brighter = cv2.add(img_rgb, matrix)
img_rgb_darker = cv2.subtract(img_rgb, matrix)
#show images
plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(img_rgb_darker);plt.title('Darker')
plt.subplot(132); plt.imshow(img_rgb);plt.title('Original')
plt.subplot(133); plt.imshow(img_rgb_brighter);plt.title('Brighter')
```

```python
matrix1 = np.ones(img_rgb.shape, dtype="uint8")* 0.8
matrix2 = np.ones(img_rgb.shape, dtype="uint8")* 1.2

img_rgb_darker = np.uint8(cv2.multiply(np.float64(img_rgb), matrix1))
img_rgb_brighter = np.uint8(cv2.multiply(np.float64(img_rgb), matrix2))

#show images
plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(img_rgb_darker);plt.title('Lower contrast')
plt.subplot(132); plt.imshow(img_rgb);plt.title('Original')
plt.subplot(133); plt.imshow(img_rgb_brighter);plt.title('High Constrast')
```

```python
#Handling the over flow of the pixels
matrix1 = np.ones(img_rgb.shape, dtype="uint8")* 0.8
matrix2 = np.ones(img_rgb.shape, dtype="uint8")* 1.2

img_rgb_darker = np.uint8(cv2.multiply(np.float64(img_rgb), matrix1))
img_rgb_brighter = np.uint8(np.clip(cv2.multiply(np.float64(img_rgb), matrix2),0,255))

#show images
plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(img_rgb_darker);plt.title('Lower contrast')
plt.subplot(132); plt.imshow(img_rgb);plt.title('Original')
plt.subplot(133); plt.imshow(img_rgb_brighter);plt.title('High Constrast')
```

```python
img_readm = cv2.imread('music.jpg',cv2.IMREAD_GRAYSCALE)

retval, img_thresh_gbl_1 = cv2.threshold(img_readm,50, 255, cv2.THRESH_BINARY)
retval, img_thresh_gbl_2 = cv2.threshold(img_readm,130, 255, cv2.THRESH_BINARY)

# Perform adaptive thresholding
img_thresh_adp = cv2.adaptiveThreshold(img_readm, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 7)

# Show the images
plt.figure(figsize=[18,15])
plt.subplot(221); plt.imshow(img_readm,        cmap="gray");  plt.title("Original");
plt.subplot(222); plt.imshow(img_thresh_gbl_1,cmap="gray");  plt.title("Thresholded (global: 50)");
plt.subplot(223); plt.imshow(img_thresh_gbl_2,cmap="gray");  plt.title("Thresholded (global: 130)");
plt.subplot(224); plt.imshow(img_thresh_adp,  cmap="gray");  plt.title("Thresholded (adaptive)");
```

```python
img_rec = cv2.imread("rectangle.png", cv2.IMREAD_GRAYSCALE)
img_cir = cv2.imread("circle.png", cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=[18, 5])
plt.subplot(131);plt.imshow(img_rec, cmap="gray");plt.title('Rectangle');
plt.subplot(132);plt.imshow(img_cir, cmap="gray");plt.title('Circle');
print(img_rec.shape)
```

```python
# Bitwise AND operation
result = cv2.bitwise_and(img_rec, img_cir, mask=None)
plt.imshow(result, cmap='gray')
```

```python
# Bitwise OR operation
result = cv2.bitwise_or(img_rec, img_cir, mask=None)
plt.imshow(result, cmap='gray')
```

```python
# Bitwise XOR operation
result = cv2.bitwise_xor(img_rec, img_cir, mask=None)
plt.imshow(result, cmap='gray')
```

```python
# Manipulation of logo
img_bgr = cv2.imread('coca-cola.png')
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
print(img_rgb.shape)
logo_w = img_rgb.shape[0]
logo_h = img_rgb.shape[1]
```

```python
img_colorCheker_bgr = cv2.imread('color_chekker.png', cv2.IMREAD_COLOR)
img_colorCheker_rgb = cv2.cvtColor(img_colorCheker_bgr, cv2.COLOR_BGR2RGB)

aspect_ratio = logo_w/img_colorCheker_rgb.shape[1]
dim =  (logo_h,logo_w)

img_colorCheker_rgb = cv2.resize(img_colorCheker_rgb, dim, interpolation=cv2.INTER_AREA)

plt.imshow(img_colorCheker_rgb)
print(img_colorCheker_rgb.shape)
```

```python
# creating masks
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)
retval, img_mask = cv2.threshold(img_gray,127,255,cv2.THRESH_BINARY)

plt.imshow(img_mask, cmap='gray')
print(img_mask.shape)
```

```python
img_mask_inv = cv2.bitwise_not(img_mask)
plt.imshow(img_mask_inv,cmap='gray')
```

```python
img_background = cv2.bitwise_and(img_colorCheker_rgb,img_colorCheker_rgb, mask=img_mask)
plt.imshow(img_background)
```

```python
img_foreground = cv2.bitwise_and(img_rgb, img_rgb, mask=img_mask_inv)
plt.imshow(img_foreground)
```

```python
result = cv2.add(img_background, img_foreground)
plt.imshow(result)
cv2.imwrite('logo_final.png',result[:,:,::-1])
```
