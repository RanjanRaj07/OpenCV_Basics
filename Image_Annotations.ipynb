```python
image = cv2.imread('rocket (2).jpg', cv2.IMREAD_COLOR)
plt.imshow(image[:,:,::-1])
```

```python
#drawing a line on the image
imageLine = image.copy()

cv2.line(imageLine, (100,100),(250,100),(0,255,255),thickness=5, lineType=cv2.LINE_AA);
plt.imshow(imageLine[:,:,::-1])
```

```python
#drawing a Circle
imageCircle = image.copy()
cv2.circle(imageCircle, (475, 525), 30, (0,0,255),thickness=3,lineType=cv2.LINE_AA)
plt.imshow(imageCircle[:,:,::-1])
```

```python
#drawing rectangle
imageRectangle = image.copy()
cv2.rectangle(imageRectangle, (385,60), (440,270), (255,0,255), thickness=3, lineType=cv2.LINE_8)
plt.imshow(imageRectangle[:,:,::-1])
```

```python
imgText = image.copy()
text = 'PSLV 3 Jupiter RR Launch, June 19, 2050'
fontScale =     2
fontFace = cv2.FONT_HERSHEY_PLAIN
fontColor = (0, 255, 0)
fontThickness = 2

cv2.putText(imgText, text, (50,600), fontFace, fontScale, fontColor, fontThickness, cv2.LINE_AA)
plt.imshow(imgText[:,:,::-1])
```
